```python
from nltk.tokenize import word_tokenize
import re
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
    # CONVERT TO LOWERCASE - Normalisation
df1['product_description'] = df1['product_description'].apply(lambda x:x.lower())

#Segmentation
texts = []
for text in df1['product_description'] :
    texts.append(text)

#Tokenization
texts_tk = [word_tokenize(t, language='english') for t in texts ]
texts_tk

#Remove puntuation 
regex = re.compile('[%s]' % re.escape(string.punctuation)) #docs
texts_pc = []
for t in texts_tk:
    word_pc = []
    for word in t:
        new_word = regex.sub(u'',word)
        if not new_word == u'':
            word_pc.append(new_word)
    texts_pc.append(word_pc)
texts_pc

#Remove stopwords
stop_words = stopwords.words('english')
texts_sw = []
for t in texts_pc:
    words_sw = []
    for word in t:
        if not word in stop_words:
            words_sw.append(word)
    texts_sw.append(words_sw)
texts_sw

#ð â \x¦ ó¾\xc© « » â\xd¤ï¸\xf Ã\x â\xa¡  æ\x\xã\x£ã\x¦ã\x\xb å\x\xdä¸\xå¾\xcä¸\x æ\x¦é\x\xæ \xbå\xb¢
texts_c = []
for t in texts_sw:
    words_c = []
    for word in t:
        new_word = word
        if "\\" not in new_word:
            #print(word)
            words_c.append(new_word)
            #print(words_c)
    texts_c.append(words_c)
texts_c

text_final = texts_c
```

```python
df1['product_description_clean'] = text_final
df1['product_description_clean'] = df1['product_description_clean'].apply(lambda x: str(x).replace(']',' '))
df1['product_description_clean'] = df1['product_description_clean'].apply(lambda x: str(x).replace('[',' '))
df1['product_description_clean'] = df1['product_description_clean'].apply(lambda x: str(x).replace("," , ' '))
df1['product_description_clean'] = df1['product_description_clean'].apply(lambda x: str(x).replace("'" , ' '))
df1['product_description_clean'] = df1['product_description_clean'].astype('object')
df1['index'] = range((len(df1)))

#Remove digits
pattern = r'[0-9]'
df1['product_description_clean'] = df1['product_description_clean'].apply(lambda x: re.sub(pattern, '', x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import KNeighborsClassifier
```

```python
#features = ['product_description_clean','product_name','index']
features = ['product_description','product_name','index']

df1 = df1[features]
```

```python
cv = CountVectorizer()
#cv_matrix = cv.fit_transform(df1['product_description_clean'])
cv_matrix = cv.fit_transform(df1['product_description'])
cv_matrix.toarray()

cos_sim = cosine_similarity(cv_matrix)
cos_sim
```

### Since we are building a content based filtering system, we need to know the users’ likes in order to predict a similar item.

### find the most similar products

```python
def get_product_index(name):
    return df1[df1['product_name']==name]['index'].values[0]

#calculate the similarity score of each product with the input one
def generate_sim_products_scores(get_product_index):
    sim_products = list(enumerate(cos_sim[get_product_index]))
    return pd.DataFrame(sim_products , columns=['i', f'product_{get_product_index}'])

#list only the similar score >= 0.2 - indexes
def list_of_sim_products_index(get_product_index , generate_sim_products_scores):
    return generate_sim_products_scores[generate_sim_products_scores[f'product_{get_product_index}']>=0.2]

#list only the similar score >= 0.2 - names
def list_of_sim_products_name(get_product_index , generate_sim_products_scores):
    sim_index_m1 = tuple(generate_sim_products_scores[generate_sim_products_scores\
                                                      [f'product_{get_product_index}']>=0.2]['i'])
    return df1[df1['index'].isin(sim_index_m1)]['product_name']

def combine_results(list_of_sim_products_index , list_of_sim_products_name):
    return list_of_sim_products_index.join(list_of_sim_products_name)

pd.set_option('display.max_rows',None)
i = get_product_index("Hornby 2014 Catalogue")
p =generate_sim_products_scores(i)
sim_scores = list_of_sim_products_index(i,p)
sim_prd = list_of_sim_products_name(i,p)
combine_results(sim_scores,sim_prd)
```
