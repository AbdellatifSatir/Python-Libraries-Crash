### List all info of specified player

```python
def player(name):
    cols = fifa.columns
    for i,row in fifa.iterrows():
        if row['Name'] == name:
            player_info = row[:]
            for i,info in enumerate(player_info):
                print(cols[i],':',info)
    else :
        print("This name does not exist")
```
