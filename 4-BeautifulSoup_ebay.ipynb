```python
attrss={'role':'heading'}
class_="s-item__price"
prices = soup.find_all(["span","span"],[attrss,class_])

price_list = []
for price in prices:
    price = price.text
    price_list.append(price)
len(price_list)
```

```python
titles = soup.find_all("span",attrs={'role':'heading'})

titles_list =[]
for title in titles:
    title = title.string
    titles_list.append(title)
len(titles_list)
```

```python
subtitles = soup.find_all("div",class_="s-item__subtitle")

subtitles_list = []
for subtitle in subtitles:
    subtitle = subtitle.text
    subtitles_list.append(subtitle)
len(subtitles_list)
```

```python
offers = soup.find_all('span',class_="s-item__purchase-options-with-icon")

offer_list = []
for offer in offers:
    offer = offer.text
    offer_list.append(offer)
len(offer_list)  # -1 missing
```

```python
shipfroms = soup.find_all("span",class_="s-item__location s-item__itemLocation")

shipfrom_list = []
for shipfrom in shipfroms:
    shipfrom = shipfrom.text
    shipfrom_list.append(shipfrom)
len(shipfrom_list)  # -1 missing
```

```python
class="s-item__info clearfix" a href
```

```python
links = soup.find_all('div',class_='s-item__info clearfix')
link_list = []
for link in links:
    link = link.a['href']
    link_list.append(link)
len(link_list)
```

### Create data set with merge , join ,concat

```python
import pandas as pd
```

```python
columns = ['Titles','Subtitles','Price','Offer','Ship','Link']
```

```python
columns[0]
```

```python
df0 = pd.DataFrame(columns=[columns[0]],data=titles_list)
df1 = pd.DataFrame(columns=[columns[1]],data=subtitles_list)
df2 = pd.DataFrame(columns=[columns[2]],data=price_list)
df3 = pd.DataFrame(columns=[columns[3]],data=offer_list)
df4 = pd.DataFrame(columns=[columns[4]],data=shipfrom_list)
df5 = pd.DataFrame(columns=[columns[5]],data=link_list)
```

#### Concat

```python
data_concat = pd.concat([df0,df1,df2,df3,df4,df5],axis=1)
```

```python
data_concat.isna().sum()
```

```python
data_concat
```

#### Join

```python
data_join = df0.join([df1,df2,df3,df4])
```
